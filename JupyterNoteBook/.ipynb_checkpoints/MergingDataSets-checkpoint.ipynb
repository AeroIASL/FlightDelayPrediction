## Import Pandas and Set Viewing Options: 

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
```

### Import FAA and BTS Datasets:

```python
FAA = pd.read_csv("../Data/FAAData/preprocessed-Jan-2018.csv")
BTS = pd.read_csv("../Data/BTSData/January/Jan-2018.csv")
```

## SFO 01/01/2018 FAA Data:

```python
FAA_SFO_01_01_2018 = FAA.loc[((FAA['Facility'] == 'SFO')&(FAA['Date'] == "01/01/2018"))]
FAA_SFO_01_01_2018
```

## 01/01/2018 BTS Data Departing From SFO :

```python
BTS_departing_SFO_01_01_2018 = BTS.loc[BTS.loc[((BTS["FL_DATE"] == "2018-01-01")&(BTS["ORIGIN"] == "SFO"))]["DEP_TIME"].sort_values().index]
BTS_departing_SFO_01_01_2018
```

## Number of Departures Are Not Alligned In Datasets:

#### BTS Departures:

```python
hours = [hour for hour in range(0,2400,100)]

departures_per_hour = []
for hour in hours:
    BTS_SFO_01_01_2018_hour_df = BTS.loc[BTS.loc[((BTS["FL_DATE"] == "2018-01-01")&(BTS["ORIGIN"] == "SFO")&(BTS["DEP_TIME"] > hour)&(BTS["DEP_TIME"] < hour+100))]["DEP_TIME"].sort_values().index]
    BTS_number_of_SFO_flights_per_hour = len(BTS_SFO_01_01_2018_hour_df.index)
    departures_per_hour.append(BTS_number_of_SFO_flights_per_hour)

hours_series = pd.Series(hours)
departures_per_hour_series = pd.Series(departures_per_hour)

BTS_SFO_01_01_2018_Departures = pd.concat([hours_series, departures_per_hour_series], axis =1)
BTS_SFO_01_01_2018_Departures.columns = ["hour","number_of_departures"]
BTS_SFO_01_01_2018_Departures
```
